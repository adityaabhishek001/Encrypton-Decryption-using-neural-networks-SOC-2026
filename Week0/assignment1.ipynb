```python
rdf = pd.read_csv('rocket_fuel_data.csv')
```

```python
rdf.isnull().sum()
```

```python
rdf['Temperature_C'] = rdf['Temperature_C'].fillna(rdf['Temperature_C'].mean())
rdf['Oxygen_Level_Percent'] = rdf['Oxygen_Level_Percent'].fillna(rdf['Oxygen_Level_Percent'].mean())
rdf['Thrust_kN'] = rdf['Thrust_kN'].fillna(rdf['Thrust_kN'].mean())
```

```python
rdf['Pressure_atm'] = rdf['Pressure_atm'].fillna(rdf['Pressure_atm'].median())
```

```python
rdf.sort_values(by='Temperature_C' , ascending=True)
```

```python
rdf.sort_values(by='Efficiency_Percent' , ascending=False)
```

```python
rdf['Power_Index'] = rdf['Thrust_kN']*rdf['Efficiency_Percent']
```
