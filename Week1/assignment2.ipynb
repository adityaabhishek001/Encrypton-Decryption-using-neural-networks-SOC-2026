```python
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as f
```

```python
import kagglehub
path = kagglehub.dataset_download("uciml/iris")
print("Path to dataset files:", path)
```

```python
import pandas as pd
import os

csv_file = os.path.join(path, "Iris.csv")
df = pd.read_csv(csv_file)
```

```python
df
```

```python
df['Species'].unique()
```

```python
def func(x):
    if x == 'Iris-setosa':
        return 0
    elif x == 'Iris-versicolor':
        return 1
    elif x == 'Iris-virginica':
        return 2
```

```python
# creating dataloaders
xtrain = []
ytrain = []
xtest = []
ytest = []
for i in range(135):
   xtrain.append(np.array([df['SepalLengthCm'][i],df['SepalWidthCm'][i],df['PetalLengthCm'][i],df['PetalWidthCm'][i]]))
   ytrain.append(df['Species'][i])
ytrain = list(map(func,ytrain))
ytrain = np.array(ytrain)
xtrain = np.array(xtrain)

for i in range(135,150):
   xtest.append(np.array([df['SepalLengthCm'][i],df['SepalWidthCm'][i],df['PetalLengthCm'][i],df['PetalWidthCm'][i]]))
   ytest.append(df['Species'][i])
ytest = list(map(func,ytest))
ytest = np.array(ytest)
xtest = np.array(xtest)
```

```python
x_train = torch.tensor(xtrain, dtype=torch.float32)
y_train = torch.tensor(ytrain, dtype=torch.long) 
x_test = torch.tensor(xtest, dtype=torch.float32)
y_test = torch.tensor(ytest, dtype=torch.long) 

from torch.utils.data import TensorDataset, DataLoader

dataset = TensorDataset(x_train, y_train)
loader = DataLoader(dataset, batch_size=1, shuffle=True)

tdataset = TensorDataset(x_test, y_test)
tloader = DataLoader(dataset, batch_size=1, shuffle=False)
```

```python
class NN(nn.Module):
    def __init__(self):
        super(NN,self).__init__()
        self.l1 = nn.Linear(4,6)
        self.l2 = nn.Linear(6,3)
        self.drop = nn.Dropout(0.3)

    def forward(self,x):
        x = f.relu(self.l1(x))
        x = self.drop(x)
        x = self.l2(x)
        return x
```

```python
num_epochs=15
learning_rate=0.001
model=NN()
criterion=nn.CrossEntropyLoss()
optimizer=torch.optim.Adam(model.parameters(),lr=learning_rate)
i=1
```

```python
for epoch in range(num_epochs):
    i=0
    for xt,yt in loader:
        flowertype = model(xt)
        loss = criterion(flowertype,yt)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'Epoch [{epoch+1}/{num_epochs}]')
```

```python
correct=0
total=0
model.eval()
for xt,yt in tloader:
    out = model(xt)
    loss = criterion(out,yt)
    correct+=(out.argmax(1)==yt).sum().item()
    total+=yt.size(0)
print(f'Test Accuracy: {100*correct/total:.2f}%')
```
